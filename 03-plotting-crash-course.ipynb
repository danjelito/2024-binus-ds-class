# Introduction to Plotting with Matplotlib and Seaborn

Visualization is a crucial aspect of data analysis, as it helps convey insights and patterns in a clear and effective manner.

Matplotlib and Seaborn are two powerful libraries in Python that provide a variety of tools for creating static, animated, and interactive visualizations.

In this section, we will explore the basics of Matplotlib and Seaborn, followed by examples of various types of plots and their usage.

## Plotting with Matplotlib

Matplotlib is a versatile library for creating static plots. It allows for fine-grained control over every element of a plot.

Commonly used for simple visualizations, Matplotlib is often the foundation for other libraries like Seaborn.

## Plotting with Seaborn

Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive statistical graphics.

It simplifies many aspects of creating visualizations and is particularly useful for complex datasets.

## Various Types of Plots and Their Usage

Here, we will explore different types of plots available in Matplotlib and Seaborn, including their usage and examples.


## 1. Line Plot

**Usage:** Line plots are used to display trends over a continuous variable, typically time.

They are useful for showing relationships between two numeric variables.

**Example:**


```python
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [200, 300, 250, 400, 500, 450]

# Create a line plot
plt.plot(months, sales, marker="o")
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

## 2. Bar Chart

**Usage:** Bar charts are used to compare quantities across different categories.

They are effective for displaying categorical data.

**Example:**


```python
# Bar chart
categories = ["Product A", "Product B", "Product C"]
values = [120, 150, 90]

plt.bar(categories, values, color=["blue", "orange", "green"])
plt.title("Sales by Product")
plt.xlabel("Products")
plt.ylabel("Sales ($)")
plt.show()
```

## 3. Histogram

**Usage:** Histograms are used to represent the distribution of a numeric variable by dividing the data into bins.

They help visualize the frequency of data points in each bin.

**Example:**


```python
import numpy as np

# Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram
plt.hist(data, bins=30, color="purple", alpha=0.7)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## 4. Scatter Plot

**Usage:** Scatter plots are used to show the relationship between two continuous variables.

They can reveal correlations, clusters, or outliers.

**Example:**


```python
# Scatter plot
height = [150, 160, 165, 170, 175, 180]
weight = [50, 60, 65, 70, 75, 80]

plt.scatter(height, weight, color="red")
plt.title("Height vs. Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

## 5. Box Plot

**Usage:** Box plots are used to visualize the distribution of a dataset and highlight its median, quartiles, and outliers.

They are helpful for comparing distributions across categories.

**Example:**


```python
import seaborn as sns

# Load a sample dataset
tips = sns.load_dataset("tips")

# Create a box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

## 6. Heatmap

**Usage:** Heatmaps are used to visualize data in a matrix format, where individual values are represented as colors.

They are useful for identifying patterns and correlations in data.

**Example:**


```python
# Create a correlation matrix and a heatmap
correlation_matrix = tips.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

## 7. Violin Plot

**Usage:** Violin plots combine box plots and density plots to show the distribution of data across different categories.

They provide more information about the distribution than box plots alone.

**Example:**


```python
# Create a violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```
